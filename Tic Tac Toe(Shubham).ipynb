```python
# Declare Global Variable
board = ['','','','','','','','','']
game = True
count = 0
player_1 = True
player_2 = True
position = 0
winner = ''
```

```python
# Display Board
def dis_board():
    print(board[0],' |',board[1],' |',board[2])
    print('--*---*--')
    print(board[3],' |',board[4],' |',board[5])
    print('--*---*--')
    print(board[6],' |',board[7],' |',board[8])
```

```python
def winner_check():
    global board
    global winner
    
    # Check Row Wise
    if board[0]==board[1]==board[2] != '':
        winner = board[0]
    if board[3]==board[4]==board[5] != '':
        winner = board[3]
    if board[6]==board[7]==board[8] != '':
        winner = board[6]

    # Check Column Wise
    if board[0]==board[3]==board[6] != '':
        winner = board[0]
    if board[1]==board[4]==board[7] != '':
        winner = board[1]
    if board[2]==board[5]==board[8] != '':
        winner = board[2]

    # check_diagonal
    if board[0]==board[4]==board[8] != '':
        winner = board[0]
    if board[6]==board[4]==board[2] != '':
        winner = board[6]
        
```

```python
# Randomly pick chance of which player should go first
def player():
    
    global player_1
    global player_2
    
    from random import randint
    change = randint(1,2)
    
    if change == 1:
        player_1 = True
        player_2 = False
    elif change == 2:
        player_1 = False
        player_2 = True
```

```python
# Player Input Functions
def player_input():
    global board
    global game
    global count
    global player_1
    global player_2
    global position
    global winner
    
    player()
    
    while game:
        while player_1:
            
            # Take position for player 'X'
            position = input("Enter position of Player 1")
            while (position=='' or int(position) not in range(1,10)):
                print("You have entered the wrong number,please enter number from 1 to 9")
                position = input()
            position = int(position)
            position=position-1
            
            # Check if position is empty
            if board[position]=='':
                board[position] = 'X'
                dis_board()
                count += 1
                player_1 = False
                player_2 = True
                winner_check()    # Check for winner
                if winner == 'X':
                    player_2 = False
                    print("Winner is player 1")
                    game = False
                    break
                if count == 9:    #Tie Condition
                    print("Game is tied..")
                    player_2 = False
                    game = False
                    break
            else:
                print("Place is not empty,Please enter again")
        while player_2:
            
            # Take position for player 'O'
            position = input("Enter position of Player 2")
            while (position=='' or int(position) not in range(1,10)):
                print("You have entered the wrong number,please enter number from 1 to 9")
                position = input()
            position = int(position)
            position=position-1

            
            # Check if position is empty
            if board[position]=='':
                board[position] = 'O'
                dis_board()
                count += 1
                player_1 = True
                player_2 = False
                winner_check()    # Check for winner
                if winner == 'O':
                    player_1 = False
                    print("Winner is player 2")
                    game = False
                    break
                if count == 9:    #Tie Condition
                    print("Game is tied..")
                    player_1 = False
                    game = False
                    break                
            else:
                print("Place is not empty,please enter again")

    
```

```python
# Reset Board
def reset_game():
    global board
    global game
    global count
    global player_1
    global player_2
    global position
    global winner
    
    board = ['','','','','','','','','']
    game = True
    count = 0
    player_1 = True
    player_2 = True
    position = 0
    winner = ''
```

```python
# Start Game
def game_start():
    play_again = 1
    while(play_again==1):
        dis_board()
        player_input()
        play_again = int(input("Enter 1 to play again"))
        if (play_again == 1):
            reset_game()
        else:
            break
game_start()
```
